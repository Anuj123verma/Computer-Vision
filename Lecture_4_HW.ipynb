```
import cv2
import matplotlib.pyplot as plt
import math
import numpy as np
```

```
image = cv2.imread("/content/Blue_Winged_Warbler_0078_161889.jpg")
```

```
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```
image = cv2.resize(image,(240,240))
```

```
plt.imshow(image)
```

```
spp=[]
division = [4,9,16,25]
for i in range(len(division)):
  p =int(math.sqrt(division[i]))
  nof = 240//p
  prev_row = 0
  curr_row = nof
  for j in range(p):
    prev_col=0
    curr_col=nof
    for k in range(p):
      sub_matrix= image[prev_row:curr_row, prev_col:curr_col]
      r = sub_matrix[:,:,0]
      g = sub_matrix[:,:,1]
      b = sub_matrix[:,:,2]
      m1 = np.mean(r)
      m2 = np.mean(g)
      m3 = np.mean(b)
      s1 = np.std(r)
      s2 = np.std(g)
      s3 = np.std(b)
      spp.extend([m1,m2,m3,s1,s2,s3])
      prev_col = curr_col
      curr_col = curr_col+nof
    prev_row=curr_row
    curr_row=curr_row+nof
print(spp)
print(len(spp))
```
